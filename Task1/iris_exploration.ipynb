# Exploring and Visualizing a Simple Dataset (Iris)

## 1. Problem Statement and Goal
The purpose of this task is to explore and understand a simple, well-structured dataset using basic data inspection and visualization techniques. The focus is on identifying patterns, distributions, and relationships between features in order to build intuition about the data before applying any machine learning models. This step helps ensure that modeling decisions are informed by a solid understanding of the dataset.

```python
# Libraries Import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Better plots
sns.set(style="whitegrid")
```

## 2. Dataset Description
The Iris dataset was used for this task. It consists of 150 samples of iris flowers, each described by four numerical features:

- Sepal length  
- Sepal width  
- Petal length  
- Petal width  

Each sample belongs to one of three species: Setosa, Versicolor, or Virginica.  
The dataset was loaded using the seaborn library and processed as a pandas DataFrame for analysis.

```python
#Loading Dataset via Seaborn
df = sns.load_dataset("iris")
```

## 3. Dataset Loading and Inspection
After loading the dataset into a pandas DataFrame, an initial inspection was performed to understand its structure and quality.

The following steps were carried out:
- Verified the dataset shape and column names  
- Displayed the first few rows using `df.head()`  
- Used `df.info()` to check data types and confirm the absence of missing values  
- Used `df.describe()` to review summary statistics such as mean, standard deviation, and value ranges  

From this inspection, it was observed that:
- The dataset contains 150 rows and 5 columns  
- There are no missing values  
- All feature columns are numerical  
- Feature ranges are relatively similar, suggesting that scaling is not immediately critical

```python
# Dataset Shape
print("Shape:", df.shape)
# Column names
print("Columns:", df.columns.tolist())
# First 5 rows
df.head()
```

```python
#checks structural validity
df.info()
#checks statistical sanity
df.describe()
```

## 4. Data Exploration and Visualization
To gain deeper insight into the dataset, multiple visualizations were created using matplotlib and seaborn.

### 4.1 Scatter Plots (Feature Relationships)
Scatter plots were used to explore relationships between different feature pairs and to observe how samples from different species are distributed. It was observed that petal-related features provide clearer separation between species compared to sepal-related features. In particular, the Setosa species appears well-separated from the others.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species"
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

### 4.2 Histograms (Feature Distributions)
Histograms were generated for each numerical feature to examine their distributions. These plots showed that most features follow reasonably smooth distributions. Petal features exhibit greater variability across species, and no extreme skewness or abnormal distributions were present.

```python
df.hist(figsize=(10,6))
plt.suptitle("Feature Distributions", y=1.02)
plt.show()
```

### 4.3 Box Plots (Outlier Detection)
Box plots were used to identify potential outliers and compare feature ranges. From these visualizations, no significant outliers were detected that would require removal. Petal length and petal width again showed clearer distinctions between species.

```python
#to deal with outliers
plt.figure(figsize=(10,5))
sns.boxplot(data=df.drop(columns="species"))
plt.title("Box Plot of Features")
plt.show()
```

### 4.4 Pair Plot (Overall Feature Comparison)
A pair plot was created to visualize all pairwise feature relationships simultaneously, with color coding based on species. This visualization clearly showed that petal length and petal width offer the strongest class separation. Setosa is the easiest species to distinguish, while Versicolor and Virginica overlap significantly and are therefore harder to separate.

```python
#to see data rather then numerics(for small datasets)
sns.pairplot(df, hue="species")
plt.show()
```

Petal-based features are more informative for species classification than sepal-based features. Visual patterns suggest that even simple classification models should perform well.
